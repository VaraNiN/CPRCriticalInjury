```python
# THIS IS AMATEUR CODE - USE AT YOUR OWN RISK!!!
# DO NOT EXECUTE CODE AN INTERNET STRANGER GAVE YOU WITHOUT READING WHAT IT DOES FIRST!
# THEREFORE, PLEASE DO SO! I AM NOT RESPONSIBLE IF YOUR COMPUTER BREAKS!

import numpy as np
import numpy.random as rng


ToHit = 2           # D10 roll that is needed to hit the enemy
Iterations = 1e5    # Number of iterations for the simulation. I suggest not going over 1e7. Python loops are slow lol


Dice = [1, 2, 3, 4, 5, 6]
Iterations = int(Iterations)
Testset = rng.randint(1, 7, (Iterations, 5))
CritRAW = np.zeros(4)
CritHOME = np.zeros(4)
for i in range(Iterations):
    Count = np.zeros((4, 6))
    for j in range(5):
        if Testset[i, j] in Dice:
            Count[0, Testset[i, j] - 1] += 1
            if j < 4:
                Count[1, Testset[i, j] - 1] += 1
            if j < 3:
                Count[2, Testset[i, j] - 1] += 1
            if j < 2:
                Count[3, Testset[i, j] - 1] += 1
    for k in range(4):
        if Count[k, 5] >= 2:
            CritRAW[k] += 1
        if any(c >= 2 for c in Count[k, :]):
            CritHOME[k] += 1

            
            
CritRAW10 = CritRAW * (100./Iterations)   # Analytical Solution: (1 - ((5/6)^x+x*(1/6)*(5/6)^(x-1))) where x is x in xd6   
CritHOME10 = CritHOME * (100./Iterations)  
CritRAWHit = ((10 - ToHit) * CritRAW + CritRAW) * (10./Iterations)      # = CritRAW10 * (10-ToHit+1)/10
CritHOMEHit = ((10 - ToHit) * CritRAW + CritHOME) * (10./Iterations) 



print("RAW, the chance to get a critical injury before rolling is (assuming a roll of %i hits):" %ToHit)
print("2d6: %.2f%%" %(CritRAWHit[3]))
print("3d6: %.2f%%" %(CritRAWHit[2]))
print("4d6: %.2f%%" %(CritRAWHit[1]))
print("5d6: %.2f%%" %(CritRAWHit[0]))

print("\nHomeruled, the chance to get a critical injury before rolling is (assuming a roll of %i hits):" %ToHit)
print("2d6: %.2f%%" %(CritHOMEHit[3]))
print("3d6: %.2f%%" %(CritHOMEHit[2]))
print("4d6: %.2f%%" %(CritHOMEHit[1]))
print("5d6: %.2f%%" %(CritHOMEHit[0]))

print("\n\n\nRAW, the chance to get a critical injury on a 10 is:")
print("2d6: %.2f%%" %(CritRAW10[3]))
print("3d6: %.2f%%" %(CritRAW10[2]))
print("4d6: %.2f%%" %(CritRAW10[1]))
print("5d6: %.2f%%" %(CritRAW10[0]))

print("\nHomeruled, the chance to get a critical injury on a 10 is:")
print("2d6: %.2f%%" %(CritHOME10[3]))
print("3d6: %.2f%%" %(CritHOME10[2]))
print("4d6: %.2f%%" %(CritHOME10[1]))
print("5d6: %.2f%%" %(CritHOME10[0]))
```
